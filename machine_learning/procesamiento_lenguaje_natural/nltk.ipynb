```python
import nltk
```

```python
nltk.download('cess_esp')
```

```python
import re
corpus = nltk.corpus.cess_esp.sents()
print(corpus)
```

```python
print(len(corpus))  # Número de oraciones en el corpus
```

```python
flatten = [w for l in corpus for w in l]
print(len(flatten))  # Número de palabras en el corpus
```

```python
print(flatten[:20])
```

```python
print(flatten[:20])
```

```python
arr = [w for w in flatten if re.search('es', w)]
```

```python
print(arr[:5])
```

# Expresiones regulares

```python
arr = [w for w in flatten if re.search('es$', w)]
print(arr[:5])
```

```python
arr = [w for w in flatten if re.search('^es', w)]
print(arr[:5])
```

```python
# Rango [a-z] [ghi]
```

```python
arr = [w for w in flatten if re.search('^[ghi]', w)]
print(arr[:10])
```

```python
# Clausura
# * repetir 0 o mas veces
# + repetir 1 o mas veces
# ? 0 o 1 vez

arr = [w for w in flatten if re.search('^(no)+', w)]
print(arr[:10])
```

```python
arr = [w for w in flatten if re.search('^(no)*', w)]
print(arr[:10])
```

# Normalizacion de texto (como aplicacion de las expresiones regulares)

#### *Tokenizacion:* Es el proceso mediante el cual se sub-divide una cadena de texto en unidades linguisticas minimas (palabras)

```python
# raw
print(r'Esta es \n una prueba')
```

```python
texto = """ Cuando sea el rey del mundo (imaginaba el en su cabeza) no tendré que preocuparme por estas bobadas.
            Era solo un niño de 7 años, pero pensaba que podria ser cualquier cosa que su imaginalción le permitiera visualizar en su cabeza."""
```

```python
# Caso 1: tokenizar por espacios vacios
print(re.split(r' ', texto))
```

```python
# casi 2: Tokenizar usando regex
print(re.split(r'[ \t\n]+', texto))
```

```python
# caso 3: 
print(re.split(r'[ \W\t\n]+', texto))
```

# Tokenizador de NLTK

```python
texto = "En los E.U. esa postal vale $15.50 ..."
print(re.split(r'[ \W\t\n]+', texto))
```

```python
pattern = r'''(?x)                  
              (?:[A-Z]\.)+            
              | \w+(?:-\w+)*         
              | \$?\d+(?:\.\d+)?%?  
              | \.\.\.              
              | [][.,;&quot;'?():-_`]    
'''
```

```python
nltk.regexp_tokenize(texto, pattern)
```
