```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def step_function(x):
    return np.piecewise(x, [x < 0, x >= 0], [0, 1])
```

```python
x = np.linspace(10, -10, 100)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x
```

```python
plt.plot(x, sigmoid(x))
```

```python
plt.plot(x, step_function(x))
```

```python
plt.plot(x, relu(x))
```

# MSE

```python
def mse(y, y_hat, derivate=False):
    if derivate:
        return (y_hat - y)
    else:
        return np.mean((y - y_hat) ** 2)
```

```python
real = np.array([0, 0, 1, 1])
pred = np.array([0.9, 0.5, 0.2, 0.0])
```

```python
mse(real, pred)
```
