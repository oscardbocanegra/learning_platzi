## Funcion lineal

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100

m = 10

b = 5

def f(x):
    return m*x+b

x = np.linspace(-10,10, N)

x
```

```python
y = f(x)

fig, ax = plt.subplots()
plt.plot(x,y)
plt.grid()
```

## Funciones polinomicas

```python
def f(x):
    return (2*x**7) - (x**4) + (3*x**2)+4

y = f(x)

plt.plot(x,y)
plt.grid()

```
