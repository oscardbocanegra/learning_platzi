```python
import numpy as np
import matplotlib.pyplot as plt
```

Funciones Lineales

```python
N = 100

m = 1

b = 5

def f(x):
    return m * x + b

x = np.linspace(-10,10, num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```

Funciones polinomicas

```python
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x, y)
plt.grid()
```

Funciones tracendentes

```python
def f(x):
    return np.cos(x)

y = f(x)

plt.plot(x, y)
plt.grid()
```

Funcion exponencial
On the following case we can see the euler number is call by the method np.exp

```python
def f(x):
    return np.exp(x)

y = f(x)

plt.plot(x, y)
plt.grid()
```

Funcion logaritmo

```python
def f(x):
    return np.log2(x)

x = np.linspace(0.001, 256, num=1000)

plt.plot(x, f(x))
plt.grid()
```

Funcion seccionada

```python
def H(x):
    Y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x>=0:
            Y[idx] = 1
    return Y

N = 1000

x = np.linspace(-10, 10, num=N)

y = H(x)

plt.plot(x, y)
plt.grid()
```
