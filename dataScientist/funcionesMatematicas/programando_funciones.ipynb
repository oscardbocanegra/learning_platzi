```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100

m = 1

b = 5

def f(x):
    return m * x + b

x = np.linspace(-10,10, num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```
