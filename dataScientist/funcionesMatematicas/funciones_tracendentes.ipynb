Funciones tracendentes 

Son funciones que no pueden ser expresadas con polinomios.

```python
#funciones trigonometricas
import numpy as np
import matplotlib.pyplot as plt

N = 1000

m = -1

b = 3

x = np.linspace(-10,10, N)  

def f(x):
    return np.cos(x)

y = f(x)

plt.plot(x,y)
```

```python
#Funcion exponencial

def f(x):
    return np.exp(x)

y=f(x)
plt.plot(x,y)
```
