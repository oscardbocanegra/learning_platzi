```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

Now if we can add diferent information to the graphics we can use the following commands

```python
#If we want to configurate the size of the iamge we can use the figsize
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y, label="sin(x)")
#With the following command we add the title to the graphic
axes[0].set_title("$Relacion X-Y$")
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend(loc="lower left")

axes[1].plot(y,x)
axes[1].set_title('Relacion Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```
