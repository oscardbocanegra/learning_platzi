```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

Now we will create 2 differents distributions in where the first one is the graphic about (x,y) and the second one is the graphic (y,x) 

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```
