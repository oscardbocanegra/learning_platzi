```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'y')
axes2.set_facecolor('gray')

fig.show()
```

As we can see on the last graphic we difined 2 tables in where we choose in whare places, dimetions and all the thigs that we can difined
