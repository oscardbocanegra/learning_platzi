```python
import matplotlib.pyplot as plt
import numpy as np
```

Now if we can generate a bar graphic we can use the following structure

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population, width=0.5, color=['red', 'blue', 'yellow', 'pink', 'green'])
#if w ewant to drive our graphics we can do the next steps
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'), rotation=45)
plt.title("population by country")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```
