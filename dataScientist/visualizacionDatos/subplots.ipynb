```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes= plt.subplots()
axes.plot(x,y)
```

```python
fig, axes= plt.subplots(nrows=1, ncols=2)
```

On the last iamge we created an figure and axes and now if we want to acces to a those grphics we will use the following commands

```python
fig, axes= plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

Now if we want to defind diferents axes we can join to then wit the following steps

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x), 'y')
axes[0,2].plot(x,np.tan(x), 'r')
axes[0,3].plot(x,np.cos(x)**2, 'g')

axes[1,0].plot(x,np.cos(x)**2, 'g')
axes[1,1].plot(x,np.sin(x)**2, 'y--')
axes[1,2].plot(x,np.cos(x), 'bo--')
axes[1,3].plot(x,np.tan(x)**2, 'rD--')
fig.tight_layout()
```
