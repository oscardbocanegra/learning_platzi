### Visualizations in Jupyter Notebook

Visualizations are a powerful way to represent data graphically, making it easier to identify patterns, trends, and insights. In Jupyter Notebook, visualizations are typically created using Python libraries. Below are some key points about visualizations:

#### Common Libraries for Visualizations
1. **Matplotlib**: A versatile library for creating static, animated, and interactive plots.
2. **Seaborn**: Built on top of Matplotlib, it provides a high-level interface for drawing attractive statistical graphics.
3. **Plotly**: A library for creating interactive plots, including 3D visualizations.
4. **Bokeh**: Focuses on creating interactive and web-ready visualizations.
5. **Pandas**: Includes built-in plotting capabilities for quick visualizations of data.

#### Types of Visualizations
1. **Line Plot**: Used to display trends over time or continuous data.
2. **Bar Chart**: Ideal for comparing categories or groups.
3. **Histogram**: Displays the distribution of a dataset.
4. **Scatter Plot**: Shows relationships or correlations between two variables.
5. **Box Plot**: Summarizes the distribution of data and highlights outliers.
6. **Heatmap**: Represents data in a matrix format with color coding.
7. **Pie Chart**: Displays proportions of a whole.

#### Steps to Create Visualizations
1. Import the required library (e.g., `import matplotlib.pyplot as plt`).
2. Prepare the data for visualization.
3. Use the appropriate plotting function (e.g., `plt.plot()` for a line plot).
4. Customize the plot (e.g., add titles, labels, legends).
5. Display the plot using `plt.show()` or save it to a file.

#### Example Code
```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y, marker='o', label='Linear Growth')
plt.title('Example Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

#### Interactive Visualizations
For interactive visualizations, libraries like Plotly and Bokeh allow users to zoom, pan, and hover over data points for more details. These are especially useful for large datasets or dashboards.

By leveraging these tools, you can effectively communicate insights and make data-driven decisions.

````python
import pandas as pd
import seaborn as sns

iris = sns.load_dataset("iris")

iris.head()
````

### Scatter Plot

A **scatter plot** is a type of data visualization that displays values for two variables as points on a two-dimensional graph. It is commonly used to observe relationships, patterns, or correlations between two continuous variables.

#### Key Features of a Scatter Plot:
1. **X-axis and Y-axis**: Each axis represents one of the variables being compared.
2. **Data Points**: Each point represents an observation in the dataset, with its position determined by the values of the two variables.
3. **Correlation**: Scatter plots help identify the type of relationship between variables:
    - **Positive Correlation**: As one variable increases, the other also increases.
    - **Negative Correlation**: As one variable increases, the other decreases.
    - **No Correlation**: No discernible relationship between the variables.

#### Use Cases:
- Identifying trends or patterns in data.
- Detecting outliers or anomalies.
- Exploring relationships between variables.

#### Customizations:
- **Color (`hue`)**: Differentiate points based on a categorical variable.
- **Size (`size`)**: Adjust the size of points based on another variable.
- **Style (`style`)**: Use different markers for categories.

Scatter plots are a simple yet powerful tool for visualizing relationships in data.

````python
# scatterplot
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
````

````python
# joinplot
sns.jointplot(data=iris, x="sepal_length", y="petal_length", hue="species")
````

````python
#box plot
sns.boxplot(x="species", y="sepal_length", data=iris)
````

````python
# barplot
sns.barplot(x='species', y='sepal_length', data= iris )
````
