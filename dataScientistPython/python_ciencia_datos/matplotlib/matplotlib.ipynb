```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
month = np.array(['E', 'F', 'M', 'A', 'MA', 'J',])
sales = np.array([20, 25, 30, 28, 35, 24])

# COnfigurar el tama√±o del grafico
plt.figure(figsize=(8, 6))

#Crear el grafico
plt.plot(month, sales, marker='o', color='blue')

plt.title('Ventas por mes')
plt.xlabel('Mes')
plt.ylabel('Ventas en miles de unidades')

plt.show()
```

Grafico de dispercion

```python
#Configurar el tama√±o del grafico
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8]
exam = [20, 25, 30, 28, 35, 24, 42]

plt.scatter(hours, exam, color='red')

plt.title('Relacion entre horas estudiadas y el puntaje')
plt.xlabel('Horas estudiadas')
plt.ylabel('Puntaje obtenido')  

plt.show()
```

Personalizaci√≥n de Gr√°ficos

```python
#Configurar el tama√±o del grafico
plt.figure(figsize=(8, 6))

hours = [2, 3, 4, 5, 6, 7, 8]
exam_scores_student_1 = [20, 25, 30, 28, 35, 24, 42]
exam_scores_student_2 = [43, 50, 35, 40, 23, 37, 50]

#Crear el grafico de dispersion de 2 estudiantes
plt.scatter(hours, exam_scores_student_1,marker='o' , color='green', linestyle='-', label='Estudiante 1')
plt.scatter(hours, exam_scores_student_2,marker='s' , color='red', linestyle='--', label='Estudiante 2')


plt.title('Relacion entre horas estudiadas y el puntaje de dos estudiantes')
plt.xlabel('Horas estudiadas')
plt.ylabel('Puntaje obtenido')  

plt.show()
```

Gr√°ficos de Barras y Diagramas de Pastel

```python
plt.figure(figsize=(10, 8))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [120, 150, 90]

#Creacion del grafico de barras
plt.bar(categories, sales, label = 'Ventas mensuales', color=['red', 'blue', 'green'])

#Anotacion con flecha para destacar un punto especifico
plt.annotate('Maximo de ventas', xy=('Producto B', 150),xytext=('Producto C', 160) ,arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Ventas de productos')
plt.xlabel('Productos')
plt.ylabel('Ventas')

plt.legend()
plt.show()
```

```python
# Configurar el tama√±o de la figura
plt.figure(figsize=(8, 6))

# Datos de ejemplo_ Participacion de mercado por producto
categories = ['Producto A', 'Producto B', 'Producto C']
sales = [50, 35, 15]

#Creacion del diagrama de pastel
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'orange'])

#Titulo
plt.title('Participacion de mercado por producto')

# Asegurar que el grafico sea un circulo
plt.axis('equal')

plt.show()
```

```python
data = np.random.normal(170, 10, 200)
print(data)
```

```python
plt.hist(data, color='salmon',bins=10 , edgecolor='black', alpha=0.6)
plt.title('Distribucion de la altura')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')
plt.show()
```

```python
np.random.seed(0)
ages = [
    np.random.normal(30, 5, 100),
    np.random.normal(40, 5, 100),
    np.random.normal(35, 5, 100)
]
print(ages)
```

```python
plt.boxplot(ages, patch_artist=True, notch=True, label=['Grupo 1', 'Grupo 2', 'Grupo 3'])

plt.title('Distribucion de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (a√±os)')

plt.show()
```
