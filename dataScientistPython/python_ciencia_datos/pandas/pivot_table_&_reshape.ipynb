Pivot table y Reshape

```python
import pandas as pd

path = './retail.csv'
df = pd.read_csv(path, encoding='ISO-8859-1')
```

```python
print(df.head())
```

```python
# Convertir la columna InvoiceDate a tipo datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Crear una nueva columna 'Sales' que multiplica la cantidad por el precio unitario
df['Sales'] = df['Quantity'] * df['UnitPrice']

# Agrupar los datos por fecha y sumar las ventas
sales_data = df.groupby(df['InvoiceDate'].dt.date)['Sales'].sum().reset_index()

# Renombrar las columnas para mayor claridad
sales_data.columns = ['Date', 'Total Sales']

print(sales_data.head())
```

```python
pivot_table = pd.pivot_table(df, values='Quantity', index='Country', columns='StockCode', aggfunc='sum')
print(pivot_table)
```
