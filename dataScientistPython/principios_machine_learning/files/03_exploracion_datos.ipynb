```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# cargar dataset
df_partidos = pd.read_csv('../data/partidos_cebollitas.csv')
df_partidos.head(6)
```

```python
# general stadistics
df_partidos.describe()
```

```python
# Average goals for Cebollitas as Local and Visitor
goles_local_cebo = df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goles_visitante_cebo = df_partidos[df_partidos['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()

print(f"average goals for Cebollitas as Local: {goles_local_cebo}")
print(f"average goals for Cebollitas as Visitor: {goles_visitante_cebo}")
```

```python
# histogram of goals scored by Cebollitas FC (local and visitor)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df_partidos['goles_local'], kde=True, ax=ax[0], bins= 10)
ax[0].set_title('Goals Scored by Cebollitas FC as Local')

sns.histplot(df_partidos['goles_visitante'], kde=True, ax=ax[1], bins= 10, color='orange')
ax[1].set_title('Goals Scored by Cebollitas FC as Visitor')

plt.show()
```

```python
# Boxplot of goals scored by Cebollitas FC (local)
sns.boxplot(x=df_partidos['goles_local'])
plt.title('Boxplot of Goals Scored by Cebollitas FC as Local')
plt.show()
```
