## Debugging operaciones con tensores

Cuando trabajamos con tensores y las operaciones resultan no validas tendremos estos tres problemas mas comunes:
1. El tamaño o forma
2. El datatype
3. El device (dispositivo) en la que se encuentra el tensor

La forma (shape) te dice como estan organizados los elementos dentro del tensor

```python
import torch
print(torch.__version__)
```

```python
scalar = torch.tensor(2.0)
vector = torch.tensor([1.0, 2.0, 3.0])
matrix = torch.tensor([[1, 2, 3], [4, 5, 6]])
t5 = torch.randn(5, 2, 3)
```

```python
print(f"La shape de la matriz es {matrix.shape}")
print(f"La shape de t5 es {t5.shape}")
```

Tambien podemos conocer la dimension de un tensor

```python
print(f"La shape de la matriz es {matrix.ndim}")
print(f"La shape de t5 es {t5.ndim}")
```

```python
print(f"El tensor matriz tiene elementos de tipo: {matrix.dtype}")
```

```python
matriz_float32 = torch.tensor([[3.1, 3.2], [3.3, 3.4]])
matriz_uint64 = torch.tensor([[3, 3], [3, 3]])

# matriz_float32.dtype, matriz_uint64.dtype

(matriz_float32 + matriz_uint64).dtype
```

Si es necesario cambiar el tipo de dato lo podemos realizar de la siguiente manera

```python
matriz_uint64.to(torch.float)
```

Tenemos que tener en cuenta el dispositivo para que nuestro tensor este preparado. No podmeos operar con un tensor diseñado para GPU (CUDA) y uno para CPU

```python
matriz_uint64.device
```

```python
torch.cuda.is_available()
```

```python
if torch.cuda.is_available():
    matriz_unit64_cuda = matriz_uint64.to(torch.device("cuda"))

    print(matriz_unit64_cuda, matriz_unit64_cuda.type())
    print(matriz_unit64_cuda.to("cpu", torch.float32))
```

### Operaciones con tensores


Primero se crean algunos tensores en PyTorch:

```python
# Crear un tensor de ceros con un shape de (3, 4)
zeros_tensor = torch.zeros((3, 4))

# Crear un tensor de unos con un shape de (3, 4)
ones_tensor = torch.ones((3, 4))

# Crear un tensor de valores aleatorios con un shape de (2, 2)
random_tensor = torch.randn((4))
```

```python
random_tensor
```

ya con estos datos creados se realizan operaciones "element-wise"

```python
# add two tensors element-twise
added_tensor = zeros_tensor + ones_tensor

# subtract two tensors element-twise
substracted_tensor = zeros_tensor - ones_tensor

# multiply two tensors element-wise
multiplied_tensor = zeros_tensor * ones_tensor

# divide two tensos element-wise
divided_tensor = random_tensor / ones_tensor
```

```python
divided_tensor
```

Multiplicacion de matrices

```python
matrix1 = torch.randn(2, 3)
matrix2 = torch.randn(3, 2)

print(f"Matrix1 shape: {matrix1.shape}")
print(f"Matrix2 shape: {matrix2.shape}")

torch.matmul(matrix1, matrix2).shape
```
