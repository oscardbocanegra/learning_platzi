```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Cargar dataset preparado
df_partidos = pd.read_csv("data/partidos_cebollitas.csv")
df_partidos.head()
```

```python
# Aplicar la normalizacion a tiros al arco (aprox. 0-15 tiros)
scaler_norm = MinMaxScaler()

df_partidos['tiros_arco_local_norm'] = scaler_norm.fit_transform(df_partidos[['tiros_arco_local']])
df_partidos['tiros_arco_visitante'] = scaler_norm.fit_transform(df_partidos[['tiros_arco_visitante']])

#ver resultados
df_partidos[['tiros_arco_local', 'tiros_arco_visitante']].head(9)
```

```python
scaler_std = StandardScaler()

df_partidos['posesion_local_std'] = scaler_std.fit_transform(df_partidos[['posesion_local (%)']])
df_partidos['posesion_visitante_std'] = scaler_std.fit_transform(df_partidos[['posesion_visitante (%)']])

# Ver resultados
print(df_partidos[['posesion_local (%)', 'posesion_local_std']].head())
print(df_partidos[['posesion_visitante (%)', 'posesion_local_std']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Visualizar tiros al arco (Normalizacion)
sns.histplot(df_partidos['tiros_arco_local_norm'], bins = 10, ax = ax[0])
ax[0].set_title("Tiros al arco de local (Normalizado)")

# Visualizar posesion local (Estandarizacion)
sns.histplot(df_partidos['posesion_local_std'], bins = 10, ax = ax[1], color='orange')
ax[1].set_title("Posesion local (Estandarizado)")

plt.show()
```
