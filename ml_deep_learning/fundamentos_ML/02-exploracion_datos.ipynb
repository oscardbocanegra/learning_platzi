```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataset partidos
df_partidos = pd.read_csv("data/partidos_cebollitas.csv")

df_partidos.head()
```

```python
# estadisticas descriptivas generales
df_partidos.describe()
```

```python
# Promedio de goles cebollita como local vs visitante
goles_local_cebo = df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goles_visitante_cebo = df_partidos[df_partidos['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()

print('Goles en promedio de cebollitas como local: ', goles_local_cebo)
print('Goles en promedio de cebollitas como visitante: ', goles_visitante_cebo)
```

```python
# Histograma de goles marcados (local y visitante)
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df_partidos['goles_local'], kde = True, ax = ax[0], bins = 10)
ax[0].set_title('Distribucion goles quipo local')

sns.histplot(df_partidos['goles_visitante'], kde = True, ax = ax[1], bins = 10, color = 'orange')
ax[0].set_title('Distribucion goles quipo visitante')
```

```python
# Boxplot para goles del equipo local
sns.boxplot(x=df_partidos['goles_local'])
plt.title('Boxplot goles equipo local')
plt.show()
```

```python
# Boxplot para goles del equipo visitante
sns.boxplot(x=df_partidos['goles_visitante'], color = 'red')
plt.title('Boxplot goles equipo visitante')
plt.show()
```

```python
# Scatterplot de posesion vs goles marcados (local)

sns.scatterplot(x = 'posesion_local (%)', y = 'goles_local', data = df_partidos)
plt.title('Relacion posesion vs goles marcados (equipo local)')
plt.show()
```

```python
# Mapa de calor para correlacion entre variables clave
plt.figure(figsize=(10, 6))
sns.heatmap(df_partidos[['goles_local', 'goles_visitante', 'posesion_local (%)', 'posesion_visitante (%)', 'tiros_arco_local']].corr(), annot=True, cmap="coolwarm")
plt.title('Mapa de calor - correlacion entre variables')
plt.show()
```
