```python
import numpy as np

# precio, calificacion
producto_a = np.array([20.0, 4.5])
print(f"Vector producto {producto_a}")
```

```python
print(f"Numero de ejes o dimensiones del producto: {producto_a.ndim}")
print(f"Forma del producto {producto_a.shape}")
print(f"Numero total de elementos del producto {producto_a.size}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, producto_a[0], producto_a[1], angles='xy', scale_units = 'xy', scale = 1, color = 'red')
plt.title("producto_a")
plt.xlabel("Precio")
plt.ylabel("calificacion")
plt.xlim(0, 25)
plt.ylim(0, 10)
plt.grid(True, alpha = 0.3)
plt.show()
```

```python
calificaciones = np.array([
    [2, 4, 4, 1],
    [3, 5, 2, 5],
    [1, 4, 3, 5],
    [5, 2, 4, 4]
])

print(calificaciones)
```

```python
print(calificaciones.ndim)
print(calificaciones.shape)
print(calificaciones.size)
```

```python
calificaciones_usuario1 = calificaciones[0]

calificaciones_pelicula3 = calificaciones[:, 2]
print(calificaciones_pelicula3)
```

```python
imagen_tensor = np.array([
    [
        [255, 0, 0],
        [0, 255, 0],
        [0, 0, 255]
    ],
    [
        [255, 125, 100],
        [200, 255, 40],
        [115, 0, 255]
    ]
])

print(imagen_tensor.ndim)
print(imagen_tensor.shape)
print(imagen_tensor.size)
```
