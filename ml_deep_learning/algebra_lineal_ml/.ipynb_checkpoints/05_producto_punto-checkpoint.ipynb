```python
import numpy as np

u = np.array([2, 3])
v = np.array([4, 1])

producto_punto_np = np.dot(u, v)
producto_punto_actual = u @ v

print(f"Resultado con np.dot(): {producto_punto_np}")
print(f"Resultado con @: {producto_punto_actual}")
```
