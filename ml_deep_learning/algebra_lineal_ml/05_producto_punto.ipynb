```python
import numpy as np

u = np.array([2, 3])
v = np.array([4, 1])

producto_punto_np = np.dot(u, v)
producto_punto_actual = u @ v

print(f"Resultado con np.dot(): {producto_punto_np}")
print(f"Resultado con @: {producto_punto_actual}")
```

```python
import gensim.downloader as api 
print("Descargando modelo de vectores de palabras...")
word_vectors = api.load("glove-wiki-gigaword-50")
print("modelo cargado!")
```

```python
rey = word_vectors['king']
hombre = word_vectors['man']
reina = word_vectors['queen']
gato = word_vectors['cat']
```

```python
def similitud_coseno(v1, v2):
    return np.dot(v1, v1) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
sim_rey_hombre = similitud_coseno(rey, hombre)
sim_rey_reina = similitud_coseno(rey, reina)
sim_rey_gato = similitud_coseno(rey, gato)
```

```python
print(f"Similitud rey y hombre: {sim_rey_hombre:.2f}")
print(f"Similitud rey y reina: {sim_rey_reina:.2f}")
print(f"Similitud rey y gato: {sim_rey_gato:.2f}")
```
