```python
import numpy as np

datos = np.array([1, 2, 3, 4, 5])
print(datos)
print(datos.shape)
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 3, 5])
v1 = np.array([3, 4])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(x, y,  label='Puntos de datos')

plt.quiver(0,0, v1[0], v1[1], angles = 'xy', scale_units = 'xy', scale= 1, color = 'red', label = 'Vector 1')
plt.title('Mi primera visualizacion')
plt.legend()
plt.grid(True, alpha = 0.3)
plt.show()
```
