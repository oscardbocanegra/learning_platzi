```python
import numpy as np

X = np.array([
    [10, 8],
    [5, 7],
    [12, 4]
])

w_fila = np.array([[0.6, 0.4]])

print(X.shape)
print(w_fila.shape)
```

```python
w_columna = w_fila.T
print(f"w_file -> {w_fila}")
print(f"w_columna-> {w_columna}")
```

```python
predicciones = X @ w_columna
```

```python
print(f"Predicciones del rendimiento de estudiantes: {predicciones}, forma de predicciones: {predicciones.shape}")
```
