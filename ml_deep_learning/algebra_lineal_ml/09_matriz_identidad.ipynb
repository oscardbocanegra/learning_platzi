```python
import numpy as np

I = np.eye(3)
print(I)
```

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(A @ I)
```

```python
M = np.array([
    [3, 1],
    [2, 2]
])
```

```python
determinante = np.linalg.det(M)
print(determinante)
```

```python
M_inversa = np.linalg.inv(M)
print(M_inversa)
```

```python
identidad = M @ M_inversa
print(np.round(identidad, 2))
```
