```python
import numpy as np

peliculas_features = np.array([
    [5, 2],
    [1, 5],
    [4, 4]
])

pesos_generos = np.array([
    [1.1, 0.4, 0.1],
    [0.2, 0.5, 1.2]
])
```

```python
print(f"Shape pelicualas features {peliculas_features.shape}")
print(f"Shape pesos generos {pesos_generos.shape}")
```

```python
scores_prediccion = peliculas_features @ pesos_generos
print(f"Shape de la matriz de prediccion {scores_prediccion.shape}")
print(f"Matriz de scores (o nivel de confianza) para cada genero: {scores_prediccion}")
```

```python
usuarios = np.array([
    [0.9, 0.2, 0.1, 0.8],
    [0.1, 0.8, 0.7, 0.2],
    [0.8, 0.7, 0.1, 0.1]
])

peliculas = np.array([
    [0.8, 0.1, 0.1, 0.9],
    [0.2, 0.9, 0.8, 0.1]
])
```

```python
print(usuarios.shape)
print(peliculas.shape)
```

```python
peliculas_transpuesta = peliculas.T
print(peliculas_transpuesta)
print(peliculas_transpuesta.shape)
```

```python
matriz_similitud = usuarios @ peliculas_transpuesta
print(matriz_similitud)
```
