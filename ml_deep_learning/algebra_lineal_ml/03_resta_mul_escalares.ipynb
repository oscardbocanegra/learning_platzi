```python
import numpy as np

calificaciones_reales = np.array([5, 2])
calificaciones_modelo = np.array([4, 4])

print(calificaciones_reales)
print(calificaciones_modelo)
```

```python
error = calificaciones_reales -calificaciones_modelo
print(error)
```

```python
porcentaje_ajuste = 0.1

ajuste = error * porcentaje_ajuste
print(ajuste)
```
